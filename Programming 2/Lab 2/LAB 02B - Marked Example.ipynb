````python
# Importing the life-saving library and the plotting library
import numpy as np
import matplotlib.pyplot as plt 

#Setting up gravity constant
g = 9.81

#Create both Vx and Vy components of Velocity
velocity_y_beginning = 7.5*np.sin((25*np.pi)/180)
velocity_x = 7.5*np.cos((25*np.pi)/180)

#Calculate the time when altitude was increasing and the total time to create an accurate plot
time_up = velocity_y_beginning/g
total_time = time_up*2

#Creating a temporary time variable (to be used as the independent variable in the plots)
time = np.linspace(0, total_time, 1000)

#Since time is constant but we need to plot it, we create an array with length of time (using ones_like() so if we alter the size of time this will change too)
velocity_x_array = velocity_x * np.ones_like(time)

#Express the changing vertical velocity (X stays constant as gravity acts vertically and no drag is considered)
velocity_y= velocity_y_beginning - g*time

#Calculate instantaneous total magnitude of velocity using Pythagoras' theorem
velocity_mag = np.sqrt(velocity_x**2 + velocity_y**2)

#Calculate the instantaneous positions using the appropriate equation
position_y = velocity_y_beginning*time - 0.5*g*(time**2)
position_x = velocity_x*time

#Using the correct formula we calculated the Maximum altitude reached and the Max distance (only the X component needed for the Max distance as Y is 0)
Max_Altitude = velocity_y_beginning*time_up - 0.5*g*time_up**2
Max_Distance = velocity_x*total_time 

#Create the figure
plt.figure(figsize=(10, 10))

# Subplot 1: Horizontal and vertical velocities over time
plt.subplot(2, 2, 1)
plt.plot(time, velocity_x_array, label='Horizontal velocity')
plt.plot(time, velocity_y, label='Vertical velocity')
plt.xlabel('Time [s]')
plt.ylabel('Velocity [m/s]')
plt.title('Horizontal and Vertical Velocity Over Time')
plt.legend()
plt.grid(True)

# Subplot 2: Magnitude of velocity over time
plt.subplot(2, 2, 2)
plt.plot(time, velocity_mag, label='Velocity magnitude')
plt.xlabel('Time [s]')
plt.ylabel('Velocity magnitude [m/s]')
plt.title('Magnitude of Velocity Over Time')
plt.legend()
plt.grid(True)

# Subplot 3: Horizontal and vertical positions (displacement) over time
plt.subplot(2, 2, 3)
plt.plot(time, position_x, label='Horizontal displacement')
plt.plot(time, position_y, label='Altitude')
plt.xlabel('Time [s]')
plt.ylabel('Position [m]')
plt.title('Horizontal Displacement and Altitude Over Time')
plt.legend()
plt.grid(True)

# Subplot 4: Altitude as a function of horizontal displacement (projectile motion)
plt.subplot(2, 2, 4)
plt.plot(position_x, position_y)
plt.xlim(0, Max_Distance)  # Limit x-axis to maximum horizontal displacement
plt.xlabel('Horizontal displacement [m]')
plt.ylabel('Altitude [m]')
plt.title('Projectile Path (Altitude vs Horizontal Distance)')
plt.grid(True)

#First save the figure and then display it (have to save before displaying so the saved png is not blank)
plt.savefig("Projectile_graphing_example.png")
plt.show()
````
